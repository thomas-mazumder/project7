```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
```

# MNIST Autoencoder

```python
# instantiate NN
layers = [{'input_dim': 64, 'output_dim': 16, 'activation': 'sigmoid'}, 
          {'input_dim': 16, 'output_dim': 64, 'activation': 'relu'}]
n = nn.NeuralNetwork(layers, lr = 0.01, seed = 0, batch_size = 10, epochs = 10, loss_function = "MSE")
```

```python
# load data
digits = load_digits()
X = digits['data']
y = digits['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# fit
train_loss, val_loss = n.fit(X_train, X_train, X_test, X_test)
```

```python
# plot losses
fig, axs = plt.subplots(1, 2, figsize=(12,4))
axs[0].plot(range(10), train_loss)
axs[0].set_title('Train loss')
axs[1].plot(range(10), val_loss)
axs[1].set_title('Validation loss')
```

```python
# visualize performance
out = n.predict(X_train)
fig, axs = plt.subplots(1, 2, figsize=(12,4))
plt.gray()
axs[0].matshow(out[0].reshape((8,8)))
axs[1].matshow(X_train[0].reshape((8,8)))
```

```python
# Quantify reconstruction performance
n._mean_squared_error(out, X_train)
```

I chose MSE loss becuase this task involves predicting the numeric, continuous brightness of pixels, rather than binary classfication. Likewise, a relu output layer lends itself more to predicting a continuous numeric variable than sigmoid does. However, I found that a sigmoid hidden layer outperformed a relu hidden layer in terms of the loss on the validation data after an equal number of epochs. Batch size, learning rate, and epoch number were also optimized wiht a brief grid search.
